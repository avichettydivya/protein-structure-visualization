```python
#Biopython with pymol program for visualization of protein structure 
#to import required modules
from Bio.PDB import PDBParser
from Bio.PDB import PDBList
import os
import requests
import py3Dmol

# Initialize PDB downloader
pdb_list = PDBList()

# Specify the PDB ID
protein=input("enter the ID of protein from PDB:")
pdb_id=protein

if len(pdb_id)==4:

    def is_valid_pdb_id(pdb_id):
        url = f"https://data.rcsb.org/rest/v1/core/entry/{pdb_id}"
        response = requests.get(url)
    
        if response.status_code == 200:
            print(f"{pdb_id} is a valid PDB ID!")
            return True
        else:
            print(f"{pdb_id} is NOT a valid PDB ID.")
            return False
        
    a=is_valid_pdb_id(pdb_id)
    if a is True:
        print("")
    
        # Download the PDB file
        pdb_file_path = pdb_list.retrieve_pdb_file(pdb_id, pdir=".", file_format="pdb")
        if os.path.exists(pdb_file_path):
            print(f"PDB file for {pdb_id} downloaded successfully!")
        else:
            print(f"Error: PDB file '{pdb_file_path}' not found.")
            
        # Print the PDB file name
        print(f"Downloaded PDB file named as: {pdb_file_path}")

        # Initialize BioPython PDB Parser
        parser = PDBParser(QUIET=True)
        structure = parser.get_structure(pdb_id,f"{pdb_file_path}")

        def get_protein_name(pdb_id):
            url = f"https://data.rcsb.org/rest/v1/core/entry/{pdb_id}"
            response = requests.get(url)
    
            if response.status_code == 200:
                data = response.json()
                protein_name = data["struct"]["title"]
                print(f"Protein Name for {pdb_id}: {protein_name}")
                return protein_name
            else:
                print(f"PDB ID {pdb_id} not found.")
                return None
                
        get_protein_name(pdb_id)

        def get_protein_info(pdb_id):
            parser = PDBParser(QUIET=True)
            structure = parser.get_structure(pdb_id,f"{pdb_file_path}")
    
            info = {
                "Structure ID": structure.id,
                "Number of Chains": len(list(structure.get_chains())),
                "Number of Residues": sum(1 for _ in structure.get_residues()),
                "Number of Atoms": sum(1 for _ in structure.get_atoms())
             }
    
            return info
        
        print(get_protein_info(pdb_id))

        # Load PDB structure
        view = py3Dmol.view(query=pdb_id, format="pdb")
        view.setStyle({"cartoon": {"color": "spectrum"}})
        view.show()
    else:
        print("Oops!!Not a valid PDB ID")
        print("Enter the valid one ")
    
else:
    print("Oops !!!!There is no such PDBID exist")
    print("Please do enter the valid PDB ID ")
    
print("successfully the program was executed")
```
